```python
row = 0
column = 0
sentences = []
word_count = 0

workbook = xlsxwriter.Workbook('bert_data.xlsx')
worksheet = workbook.add_worksheet()

df = pd.DataFrame(data)

for i in range(0,327): # range(4) = text count
    
    text = df.loc[i, 'text']
    text_class = df.loc[i, 'class']
    
    splited_text = text.split(".")
    
    for j in splited_text: # sentences in the text
        
        if( j == splited_text[-1]):
            
            text = "".join(sentences)
            
            inputs = tokenizer([text], max_length=20000, return_tensors='pt', truncation=True)
            summary_ids = model.generate(inputs['input_ids'], num_beams=10, min_length=40, max_length=60)
            result = [tokenizer.decode(g, skip_special_tokens=True, clean_up_tokenization_spaces=False) for g in summary_ids]
            
            print(result[0])
            
            worksheet.write(row, 0, result[0])
            worksheet.write(row, 1, text_class)
            row += 1
            
            sentences = []
            word_count = 0
        
        if(word_count > 100):
            
            text = "".join(sentences)
            
            inputs = tokenizer([text], max_length=20000, return_tensors='pt', truncation=True)
            summary_ids = model.generate(inputs['input_ids'], num_beams=10, min_length=40, max_length=60)
            result = [tokenizer.decode(g, skip_special_tokens=True, clean_up_tokenization_spaces=False) for g in summary_ids]
            
            print(result[0])
            
            worksheet.write(row, 0, result[0])
            worksheet.write(row, 1, text_class)
            row += 1
            
            sentences = []
            word_count = 0
            
        else:
            word_list = j.split()
            word_count += len(word_list)
            sentences.append(j)
            
workbook.close()
```
